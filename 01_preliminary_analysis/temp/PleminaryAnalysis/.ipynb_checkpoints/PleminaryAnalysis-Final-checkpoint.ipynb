```python
chestnut, _ = librosa.load('audio/wav/chestnut_winged_babbler_sample1.wav')
unknown, _ = librosa.load('audio/wav/unknown_bird_comparison.wav')

def plot_spectrogram(signal, name):
    """Compute power spectrogram with Short-Time Fourier Transform and plot result."""
    spectrogram = librosa.amplitude_to_db(librosa.stft(signal))
    plt.figure(figsize=(20, 15))
    librosa.display.specshow(spectrogram, y_axis="log")
    plt.colorbar(format="%+2.0f dB")
    plt.title(f"Log-frequency power spectrogram for {name}", fontsize = 16)
    plt.xlabel('Time (seconds)', fontsize = 12)
    plt.ylabel('Frequency (Hz)', fontsize = 12)
    plt.show()
```

```python
plot_spectrogram(chestnut, "Chestnut-Winged Babbler")
```

Above we can see the Chestnut-winged Babbler characteristic chirps.

```python
def ReadAIFF(file):
#Reads the frames from the audio clip and returns the uncompressed data
    s = aifc.open(file,'r')
    nFrames = s.getnframes()
    strSig = s.readframes(nFrames)
    return np.fromstring(strSig, np.short).byteswap()
```

```python
def plot_sbs_spectrogram(bird, non_bird):
    bird = ReadAIFF(bird)
    non_bird = ReadAIFF(non_bird)
    plt.figure(figsize = (14,4))
    ax1 = plt.subplot(121)
    params = {'Fs':50000, 'noverlap':192}
    plt.specgram(bird, **params);
    ax1.set_title('Chestnut-winged Babbler', fontsize = 12)
    ax1.set_xlabel('Time (seconds)', fontsize = 12)
    ax1.set_ylabel('Frequency (Hz)', fontsize = 12)
    cbar = plt.colorbar()
    cbar.set_label('Amplitude', fontsize = 12)
    ax2 = plt.subplot(122)
    plt.specgram(non_bird, **params);
    ax2.set_title('Unknown Bird', fontsize = 12)
    ax2.set_xlabel('Time (seconds)', fontsize = 12)
    ax2.set_ylabel('Frequency (Hz)', fontsize = 12)
    cbar = plt.colorbar()
    cbar.set_label('Amplitude', fontsize = 12)
    plt.show()
```

```python
plot_sbs_spectrogram('audio/aiff/chestnut_winged_babbler_sample2.aiff', 'audio/aiff/unknown_bird_comparison.aiff')
```

From the two plots above we can see the CWB, in the 1000hz range, signified by 7 dots. The unknown bird can be seen in a similar range but represented with a high to low Frequency and repeated serveral times.

From here and below I am attempting to use template matching to identify sounds made by a CWB. This is just a concept, as our dataset is way to small to accuratly apply an algorithm that works. 

If we had a larger dataset, I would iterate through the audio files with save_spectrogram, classifying as bird or non bird and train the model as such.

Templating builds on https://github.com/jaimeps/whale-sound-classification/blob/master/3_template_matching_tutorial/Tutorial_template_matching.ipynb.

```python
# Save image of spectrogram (we intentionally remove labels from the chart in this case)
def save_spectrogram(filename):
    sound = ReadAIFF(filename)
    fig = plt.figure(figsize = (10,6));
    params = {'NFFT':256, 'Fs':50000, 'noverlap':192};
    plot1 = plt.specgram(sound, **params);
    return fig;
```

```python
def match_template(template_to_match, image_where_to_search):
    # Read the images
    img = cv2.imread(image_where_to_search,0)
    img2 = img.copy()
    template = cv2.imread(template_to_match,0)
    w, h = template.shape[::-1]

    # Set method 
    meth = 'cv2.TM_CCOEFF_NORMED'
    img = img2.copy()
    method = eval(meth)

    # Apply template Matching
    res = cv2.matchTemplate(img,template,method)
    # Obtain vals and locs (max_lox corresponds to the best match)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
    print('Min_val=', min_val, ' Max_val=', max_val)
    # Add rectangle with match
    bottom_right = (max_loc[0] + w, max_loc[1] + h)
    cv2.rectangle(img,max_loc, bottom_right, 255, 3)

    # Plot matching and image with best match in rectangle
    plt.figure(figsize = (14,8))
    plt.subplot(121),plt.imshow(res,cmap = 'gray', interpolation='nearest')
    plt.title('Matching Result'), plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(img, cmap = 'gray', interpolation='nearest')
    plt.title('Detected Point'), plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
# Save image of whale sound
plot1 = save_spectrogram('audio/aiff/chestnut_winged_babbler_sample2.aiff');
plot1.savefig('chestnut1.png');
plt.close(plot1)
```
